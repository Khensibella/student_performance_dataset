```python
df_cleaned.replace([np.inf, -np.inf], np.nan, inplace=True)
df_cleaned.isnull().sum()
```

<h1 style="font-size: 14px; border-radius: 20px; padding: 12px; border: 4px solid olive; line-height: 1.6;">
Distribution of numeric variables</h1>

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))  

axes = axes.ravel()

for idx, col in enumerate(num_col):
    sns.histplot(x=df_cleaned[col], ax=axes[idx], color='lightgreen')
    axes[idx].grid(True, which='major', axis='y', linestyle='--', linewidth=0.7)
    axes[idx].set_title(f'Distribution of {col}')  
    
    
plt.tight_layout() 
plt.show()
```

<h1 style="font-size: 14px; border-radius: 20px; padding: 12px; border: 4px solid darkgreen; line-height: 1.6;">
    - The breaks or gaps in the 'Hours_Studied' variable may be due to respondents answering approximately. <br>
    - The variables 'Attendance' and 'Previous_Scores', which exhibit a uniform distribution, have minimal impact on the 'Exam_Score' variable.
</h1>

<h1 style="font-size: 14px; border-radius: 20px; padding: 12px; border: 4px solid olive; line-height: 1.6;">
Relationship between the numerical variables and the target variable. (by Scatterplot) </h1>

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))  

axes = axes.ravel()

for idx, col in enumerate(num_col):
    sns.scatterplot(x=df_na[col], y=df_na[target], ax=axes[idx], color='yellowgreen')  
    axes[idx].set_title(f'Relationship b/t {col} & {target}')
    plt.xlabel(col)
    plt.ylabel(target)

plt.tight_layout()  
plt.show()
```

```python
corr_col = ['Hours_Studied', 'Attendance', 'Previous_Scores', 'Exam_Score']
df_corr = df_cleaned[corr_col]

corr = df_corr.corr()

plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True, cmap='viridis')
```

<h1 style="font-size: 14px; border-radius: 20px; padding: 12px; border: 4px solid darkgreen; line-height: 1.6;">
    - The variables 'Hours_Studied' and 'Attendance' show a positive correlation with 'Exam_Hours'. However, 'Previous_Scores' appears to have little to no relationship with 'Exam_Score'. <br>
    - I guess there are many values that deviate from the overall trend. These outliers represent smart students, instances of cheating, or the effects of other variables. <br> 
    - The most explainable variable is 'Extracurricular_Activities', which could be confirmed below.</h1>

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 5))  

axes = axes.ravel()

for idx, col in enumerate(num_col):
    sns.scatterplot(x=df_na[col], y=df_na[target], ax=axes[idx], palette='PRGn', hue=df_cleaned['Extracurricular_Activities'])  
    axes[idx].set_title(f'Relationship b/t {col} & {target}')
    axes[idx].legend(loc='upper center', bbox_to_anchor=(0.5, 1.25), ncol=2, title='Extracurricular_Activities')
    plt.xlabel(col)
    plt.ylabel(target)

plt.tight_layout()  
plt.show()
```
